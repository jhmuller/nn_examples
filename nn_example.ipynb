```
import os
import sys
import datetime
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from tensorflow import keras
```

```
# code to print a functions name from within the function
def func_name(i=1):
  return sys._getframe(i).f_code.co_name

# define callbacks 
class CustomCallbacks(keras.callbacks.Callback):
    def __init__(self, train_verbosity=0,
                       epoch_verbosity=0,
                       batch_verbosity=0,
                       verbosity=0):
      super().__init__()
      self.train_verbosity = train_verbosity
      self.epoch_verbosity = epoch_verbosity
      self.batch_verbosity = batch_verbosity

    def on_train_begin(self, logs=None):
      if self.train_verbosity > 0:
            print("{0} <{1}>".format(func_name(), datetime.datetime.now()))
        #
      if len(list(logs.keys())) > 0:
          if self.train_verbosity > 0:
            for k in logs.keys():
              print("{0}: {1}".format(k, logs[k]))

    def on_train_end(self, logs=None):
      if self.train_verbosity > 0:
        print("{0} <{1}>".format(func_name(), datetime.datetime.now()))
        #
      if len(list(logs.keys())) > 0:        
        if self.train_verbosity > 0:        
          for k in logs.keys():
            print("{0}: {1}".format(k, logs[k]))

    def on_epoch_begin(self, epoch, logs=None):
      if self.epoch_verbosity > 0:
        print("{0} epoch: {1} <{2}>".format(func_name(), epoch, 
                                         datetime.datetime.now()))
        #
      if len(list(logs.keys())) > 0:
        if self.epoch_verbosity > 0:
          for k in logs.keys():
            print("{0}: {1}".format(k, logs[k]))

    def on_epoch_end(self, epoch, logs=None):
      if self.epoch_verbosity > 0:
        print("{0} epoch: {1} <{2}>".format(func_name(), epoch, 
                                         datetime.datetime.now()))
      if len(list(logs.keys())) > 0:
        if self.epoch_verbosity > 0:
          for k in logs.keys():
            print("{0}: {1}".format(k, logs[k]))          

    def on_train_batch_begin(self, batch, logs=None):
      if self.batch_verbosity > 0:
        print("{0} batch: {1} <{2}>".format(func_name(), batch, 
                                         datetime.datetime.now()))    
        #
      if len(list(logs.keys())) > 0:        
        if self.batch_verbosity > 0:
          for k in logs.keys():
            print("{0}: {1}".format(k, logs[k]))     

    def on_train_batch_end(self, batch, logs=None):
      if self.batch_verbosity > 0:
        print("{0} batch: {1} <{2}>".format(func_name(), batch, 
                                         datetime.datetime.now()))    
        #
      if len(list(logs.keys())) > 0:        
        if self.batch_verbosity > 0:
          for k in logs.keys():
            print("{0}: {1}".format(k, logs[k]))     

    def on_test_begin(self, logs=None):
      pass

    def on_test_end(self, logs=None):
      pass
    def on_predict_begin(self, logs=None):
      pass

    def on_predict_end(self, logs=None):
      pass

    def on_test_batch_begin(self, batch, logs=None):
      pass

    def on_test_batch_end(self, batch, logs=None):
      pass

    def on_predict_batch_begin(self, batch, logs=None):
      pass

    def on_predict_batch_end(self, batch, logs=None):
      pass
```

```
# run code to fit the modelfit the model
learning_rates = [0.0001, 0.001, 0.005, 0.01]

optimizers = {"SGD": tf.keras.optimizers.SGD(),
              "RMSprop": tf.keras.optimizers.RMSprop(),
              "Adam":tf.keras.optimizers.Adam(),
              }

def gen_data(npoints=10, intercept=-1, slope=2, n_start=-2):
  n_stop = n_start + npoints
  xs = np.array([float(i) for i in range(n_start, n_stop+1)], dtype=float)
  ys = np.array([slope*x + intercept  for x in xs], dtype=float)
  return xs, ys

verbose = 0
initilaizer = tf.keras.initializers.random_normal()
k_initilaizer = tf.keras.initializers.random_uniform(minval=1, maxval=3)
b_initilaizer = tf.keras.initializers.random_uniform(minval=-2, maxval=0)
epochs = 50
for npoints in (10,20,50):
  xs, ys = gen_data(npoints=npoints)
  print("\n-----npoints= {0}------".format(npoints))
  df = pd.DataFrame({"x":xs, "y":ys})

  for k in optimizers.keys():
    model = tf.keras.Sequential([keras.layers.Dense(units=1, 
                                                    activation=None,
                                                    bias_initializer=b_initilaizer,
                                                    kernel_initializer=k_initilaizer,
                                                    input_shape=[1])])
    model.summary()      
    optimizer = optimizers[k]
    print("\n---optimizer= {0}--".format(k))
    model.compile(optimizer='sgd', loss='mean_squared_error')
    mycallbacks = CustomCallbacks(epoch_verbosity=1)
    model.fit(xs, ys, epochs=epochs, batch_size=10, 
              verbose=verbose)
    pys = []
    for x in xs:
      py = model.predict([x])
      pys.append(py[0][0])

    df['py'] = pys

    fig, ax = plt.subplots()

    df.plot(ax=ax, kind='line', x='x', y='y', c='blue', marker='o', label="orig")    
    df.plot(ax=ax, kind='line', x='x', y='py', c='green', marker='x', label="pred")    

    plt.show()     
    for v in model.trainable_variables:
      print(v)    
```
